```python
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense
from tensorflow.keras.models import Model

# 입력 시퀀스의 길이
input_length = 100

# 임베딩 차원
embedding_dim = 32

# LSTM 차원
lstm_units = 64

# 입력 레이어 정의
input_layer = Input(shape=(input_length,), name='input_layer')

# 임베딩 레이어 정의
embedding_layer = Embedding(input_dim=3270, output_dim=embedding_dim,
                            input_length=input_length, name='embedding_layer')(input_layer)

# LSTM 레이어 정의
lstm_layer = LSTM(units=lstm_units, name='lstm_layer')(embedding_layer)

# 출력 레이어 정의
output_layer = Dense(units=1, activation='linear', name='output_layer')(lstm_layer)

# 모델 생성
model = Model(inputs=input_layer, outputs=output_layer)

# 모델 컴파일
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mae'])

```
